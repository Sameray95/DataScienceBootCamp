#**Python Lambda Functions**

```python
x = lambda a: a + 10
print(x(10))
```

```python
x = lambda a, b: a*b+10
print(x(2,8))
```

```python
x = lambda a,b,c: a*b*c
print(x(3,6,9))
```

```python
x = lambda a,b: a if (a>b) else b
print(max(12,21))
```

```python
x = 5

if x > 0:
    print("x is positive")

    if x % 2 == 0:
        print("x is even")
    if x % 2 != 0:
        print("x is odd")

    if x < 10:
        print("x is less than 10")
```

```python
check_x = lambda x: (f"x is positive\n" if x > 0 else "") + (f"x is even\n" if x % 2 == 0 else "") + (f"x is odd\n" if x % 2 != 0 else "") + (f"x is less than 10\n" if x < 10 else "")

result = check_x(5)
print(result)
```

```python
result = lambda x,y : f"{x} is smaller than {y}" \
if x < y else (f"{x} is greater than {y}" if x > y \
               else f"{x} is equal to {y}")

print(result(12, 11))
print(result(12, 12))
print(result(12, 13))
```
