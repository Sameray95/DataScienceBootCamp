```python
#1) How many reviewers from each nation

#a) Highest reviewers from which nationality

#b) Lowest Reviewers from which nationality


reviewers = df.groupby('Reviewer_Nationality').size()
reviewers
```

```python
#a) Highest reviewers from which nationality
country= reviewers.idxmax()
review = reviewers.max()
print(country , " : " , review)
```

```python
#a) lowest reviewers from which nationality
country= reviewers.idxmin()
review = reviewers.min()
print(country , " : " , review)
```

```python
#2) Property with best review_score overall and least score overall

# Group by 'Hotel_Name' and calculate the average review score
average_scores = df.groupby('Hotel_Name')['Reviewer_Score'].mean().round(2)
average_scores
```

```python
# Find the hotel with the highest average review score
best_hotel = average_scores.idxmax()
best_score = average_scores.max()
print(best_hotel , " : " , best_score)
```

```python

# Find the hotel with the lowest average review score
worst_hotel = average_scores.idxmin()
worst_score = average_scores.min()
print(worst_hotel , " : " , worst_score)
```

```python
#3) Do a scatter plot to show all properties over lat and long

# Fill missing latitudes and longitudes with corresponding hotel's values
df['lat'] = df.groupby('Hotel_Name')['lat'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
df['lng'] = df.groupby('Hotel_Name')['lng'].transform(lambda x: x.fillna(method='ffill').fillna(method='bfill'))

plt.figure(figsize=(10, 6))
plt.scatter(df['lng'], df['lat'], alpha=0.5, color='blue')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Properties Distribution over Latitude and Longitude')
plt.show()

```

```python
#4) Which property has highest reviews and lowest  in last 1 month of the dataset maximum date

# Convert 'days_since_review' column to datetime
df['days_since_review'] = pd.to_timedelta(df['days_since_review'])

# Find the maximum date
max_date = df['days_since_review'].max()

# Filter for reviews within the last 1 month
reviews_last_month = df[df['days_since_review'] >= max_date - pd.Timedelta(days=30)]

# Find the property with the highest reviews
highest_reviews_property = reviews_last_month[reviews_last_month['Review_Total_Positive_Word_Counts'] == reviews_last_month['Review_Total_Positive_Word_Counts'].max()]['Hotel_Name'].unique()

# Find the property with the lowest reviews
lowest_reviews_property = reviews_last_month[reviews_last_month['Review_Total_Positive_Word_Counts'] == reviews_last_month['Review_Total_Positive_Word_Counts'].min()]['Hotel_Name'].unique()

print(f"The property with the highest reviews in the last 1 month: {highest_reviews_property}")
print(f"The property with the lowest reviews in the last 1 month: {lowest_reviews_property}")
```

```python
# Get the reviews from the first row
positive_review = df.loc[0, 'Positive_Review']
negative_review = df.loc[0, 'Negative_Review']

# Combine them
combined_review = positive_review + negative_review
```

```python
combined_review
```

```python
import nltk
nltk.download('punkt')
```

```python

# Word Tokenization
word_tokens = nltk.word_tokenize(combined_review)

# Sentence Tokenization
sent_tokens = nltk.sent_tokenize(combined_review)

# Print the tokenized results
print("Word Tokens:")
print(word_tokens)

print("\nSentence Tokens:")
print(sent_tokens)
```

```python
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize
nltk.download('wordnet')
```

```python
# Tokenize the text
tokens = word_tokenize(combined_review)

# Initialize stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Perform stemming
stemmed_tokens = [stemmer.stem(word) for word in tokens]

# Perform lemmatization
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]

# Print the results
print("Stemmed Tokens:")
print(stemmed_tokens)

print("\nLemmatized Tokens:")
print(lemmatized_tokens)
```
