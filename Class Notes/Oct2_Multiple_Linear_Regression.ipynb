In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,

2 yr experience, 9 test score, 6 interview score

12 yr experience, 10 test score, 10 interview score

```python
path = '/content/hiring.csv'
df =  pd.read_csv(path)
df
```

```python
experience_mapping = {'five': 5, 'two': 2, 'seven': 7, 'three': 3, 'ten': 10, 'eleven': 11}

df['experience'] = df['experience'].map(experience_mapping)
df
```

```python
exp_median = df['experience'].median()
tstscor_median = df['test_score(out of 10)'].median()
```

```python
df['experience'] = df['experience'].fillna(exp_median)
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(tstscor_median)
df
```

```python
from sklearn import linear_model
```

```python
df = df.rename(columns={'interview_score(out of 10)': 'interview_score'})
df = df.rename(columns={'test_score(out of 10)': 'test_score'})
df = df.rename(columns={'salary($)': 'salary'})
df
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df.drop ('salary', axis ='columns'),df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[6.0	8.0	6]])
```
