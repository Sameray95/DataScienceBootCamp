```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# replacing NaN with 'unknown' in director columns
df['director'] = df['director'].fillna('unknown')
df['director'].isnull().sum()
```

```python
# replacing NaN with 'unknown' in cast columns
df['cast'] = df['cast'].fillna('unknown')
df['cast'].isnull().sum()
```

```python
# replacing NaN with 'unknown' in country columns

df['country'] = df['country'].fillna('unknown')
df['country'].isnull().sum()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

```python
df.isna().sum()
```

```python
# Separate 'date_added' into 'year_added' and 'month_added'
df['year_added'] = df['date_added'].str.split(',').str[-1].str.strip()
df['month_added'] = df['date_added'].str.split(' ').str[0]

# Convert 'year_added' to integer
df['year_added'] = df['year_added'].astype(int)
```

```python
# droping column 'date_added'
df.drop('date_added',axis=1,inplace= True)
```

```python
df
```

```python
df.drop('listed_in',axis =1,inplace =True)
```

```python
df
```

```python
# Top Directors and Cast
top_directors = df['director'].value_counts().head(10)
top_cast = df['cast'].str.split(',').explode().str.strip().value_counts().head(10)

print('Top Directors:')
print(top_directors)

print('Top Cast:')
print(top_cast)
```

```python
# Country Analysis
country_counts = df['country'].value_counts().head(10)
print('Top Countries:')
print(country_counts)
```

```python
# Year Added Analysis
year_added_counts = df['year_added'].value_counts().sort_index()
print('Year Added Counts:')
print(year_added_counts)

```

```python
# Month Added Analysis
month_added_counts = df['month_added'].value_counts()
print('Month Added Counts:')
print(month_added_counts)
```

```python

sns.countplot(data=df, x='type')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='release_year', kde=True)
plt.title('Distribution of Release Years')
plt.show()
```

```python
# Scatter plot
sns.scatterplot(data=df, x='release_year', y='type')

# Set labels and title
plt.xlabel('Release Year')
plt.ylabel('Type')
plt.title('Release Year vs Type')

plt.show()
```
