```python
for i in range (100):
  print(i)
```

```python
#list Comprehension


lst_comp1=[i for i in range (100)]
print(lst_comp1)
```

```python
for i in range(100):
  if i%3==0:
    print(i)
```

```python
#list Comprehension

lst_comp2= [i for i in range(100) if i%3==0]
print(lst_comp2)
```

```python
fruit_list=['apple', 'banana', 'cherry', 'dates', 'eggfruit', 'fig','grapes']
for i in fruit_list:
  if 'a' in i :
    print (i)
```

```python
lst_comp3= [ i for i in fruit_list if 'a' in i ]
print(lst_comp3)
```

```python
apple_fruit = ['Apple', 'Pineapple', 'Custard apple', 'Rose apple', 'Sugar apple', 'Mayapple', 'Elephant apple', 'Crabapple', 'banana', 'cherry', 'dates', 'eggfruit', 'fig','grapes']
```

```python
lst_comp4= [i for i in apple_fruit if 'apple' in i ]
print(lst_comp4)
```

```python
lst_comp4= [i for i in apple_fruit if 'apple' not in i ]
print(lst_comp4)
```

```python
apple_fruit1 = ['Apple', 'Pineapple', 'Custard apple', 'Rose apple', 'Sugar apple', 'Mayapple', 'Elephant apple', 'Crabapple', 'banana', 'cherry', 'dates', 'eggfruit', 'fig', 'grapes']

found_apple = False

for fruit in apple_fruit1:
    if 'apple' in fruit:
        found_apple = True
        break

if found_apple:
    print("Available")
else:
    print("Not Available")
```

```python
ndim2array = [[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]]

lst1=[[i[j] for i in ndim2array] for j in range(len(ndim2array[0]))]
lst1
```

```python
#dictionary Comprehension

dictcomp={i: f"Item{i}" for i in range (10)}
dictcomp
```

```python
for i in range(1000):
  print(i,':Item', +i)
```

```python
#set Comprehension
set1={1,2,3,4,5,'sam', 'eer', 'a','apple','sam',1,2,34,4,5,'eer' }
setcomp= {i for i in set1 }
print(setcomp)
```
