```python
class Employee:
  def __init__(self, emp_id, name, position, salary,age):
    self.emp_id = emp_id
    self.name = name
    self.position = position
    self.salary = salary
    self.age = age

  def display_info(self):
    print("Employee ID: " ,self.emp_id)
    print("Name: " ,self.name)
    print("Position: " ,self.position)
    print("Salary: " ,self.salary)
    print("Age: " ,self.age)
```

```python
class EmployeeManagementSystem:
    def __init__(self):
        self.employees = []

    def Add_Employee(self, employee):
        self.employees.append(employee)

    def Delete_By_ID(self, emp_id):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                self.employees.remove(employee)
                return True
        return False

    def Update_Name_By_Id(self, emp_id, new_name):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                employee.name = new_name
                return True
        return "No employee found with Employee ID " + str(emp_id)

    def Update_Salary_By_Id(self, emp_id, new_salary):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                employee.salary = new_salary
                return True
        return "No employee found with Employee ID " + str(emp_id)

    def Update_Job_Position_By_Id(self, emp_id, new_position):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                employee.jobposition = new_position
                return True
        return "No employee found with Employee ID " + str(emp_id)

    def Update_Age_By_Id(self, emp_id, new_age):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                employee.age = new_age
                return True
        return "No employee found with Employee ID " + str(emp_id)

    def Delete_By_Name(self, name):
        for employee in self.employees:
            if employee.name == name:
                self.employees.remove(employee)
                return True
        return "No employee found with name: " + name

    def Get_Details_By_Id(self, emp_id):
        for employee in self.employees:
            if employee.emp_id == emp_id:
                return "ID: " + str(employee.emp_id) + ", Name: " + employee.name + ", Salary: " + str(employee.salary) + ", Position: " + employee.jobposition + ", Age: " + str(employee.age)
        return "No employee found with Employee ID " + str(emp_id)

    def Get_Details_By_Name(self, name):
        for employee in self.employees:
            if employee.name == name:
                return "ID: " + str(employee.emp_id) + ", Name: " + employee.name + ", Salary: " + str(employee.salary) + ", Position: " + employee.jobposition + ", Age: " + str(employee.age)
        return "No employee found with name: " + name
```

```python
ems = EmployeeManagementSystem()

while True:
    print("\nMenu:")
    print("1. Add Employee")
    print("2. Update Employee")
    print("3. Delete Employee")
    print("4. Display All Employees")
    print("5. Exit")
    choice = input("Enter your choice (1/2/3/4/5): ")

    if choice == '1':
        name = input("Enter employee name: ")
        salary = float(input("Enter employee salary: "))
        age = int(input("Enter employee age: "))
        jobposition = input("Enter employee job position: ")
        employee = Employee(salary, name, age, jobposition)
        ems.Add_Employee(employee)
        print(name + " has been added as an employee.")


    elif choice == '2':
        emp_id = int(input("Enter Employee ID to update: "))
        print("1. Update Name")
        print("2. Update Salary")
        print("3. Update Job Position")
        print("4. Update Age")
        update_choice = input("Enter your choice (1/2/3/4): ")
        if update_choice == '1':
            new_name = input("Enter new name: ")
            result = ems.Update_Name_By_Id(emp_id, new_name)
            print(result)
        elif update_choice == '2':
            new_salary = float(input("Enter new salary: "))
            result = ems.Update_Salary_By_Id(emp_id, new_salary)
            print(result)
        elif update_choice == '3':
            new_position = input("Enter new job position: ")
            result = ems.Update_Job_Position_By_Id(emp_id, new_position)
            print(result)
        elif update_choice == '4':
            new_age = int(input("Enter new age: "))
            result = ems.Update_Age_By_Id(emp_id, new_age)
            print(result)

    elif choice == '3':
        emp_id = int(input("Enter Employee ID to delete: "))
        result = ems.Delete_By_ID(emp_id)
        if result:
            print("Employee with ID " + str(emp_id) + " deleted.")
        else:
            print("No employee found with ID " + str(emp_id) + ".")

    elif choice == '4':
        for employee in ems.employees:
            employee.display_info()



    elif choice == '5':
        break

    else:
        print("Invalid choice. Please enter 1, 2, 3, 4, or 5.")
```

```python
class Employee:
  def __init__(self, emp_id, name, position, salary,age):
    self.emp_id = emp_id
    self.name = name
    self.position = position
    self.salary = salary
    self.age = age

  def display_info(self):
    print("Employee ID", self.emp_id)
    print("Name", self.name)
    print("Position", self.position)
    print("Salary", self.salary)
    print("Age", self.age)
```

```python
class EmployeeManagementSystem:
  def __init__(self):
    self.employees = []

  def add_employee(self, employee):
    self.employees.append(employee)
    print(len(self.employees))

  def remove_employee(self, emp_id):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        self.employees.remove(employee)
        return True
    return False

  def update_name(self,emp_id,name):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.name= name
        return True
    return False

  def update_salary(self,emp_id,salary):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.salary= salary
        return True
    return False

  def update_position(self,emp_id,position):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.position= position
        return True
    return False

  def update_age(self,emp_id,age):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.age= age
        return True
    return False

  def delete_by_name(self,name):
    all_matching_names=[]
    for employee in self.employees:
      if employee.name == name:
        all_matching_names.append(employee)
    if len(all_matching_names)==0:
      return False
    if len(all_matching_names)==1:
      self.employees.remove(all_matching_names[0])
      return True
    if len(all_matching_names)>1:
      print("Duplicate names found please select the id you want to remove")
      all_ids=[]
      for emps in all_matching_names:
        print("ID :- ",emps.emp_id)
        all_ids.append(emps.emp_id)
      id_to_remove=int(input("Please choose the id you want to remove :- "))
      if id_to_remove in all_ids:
        self.employees.remove(all_matching_names[all_ids.index(id_to_remove)])
        return True
      else:
        return False

  def details_by_id(self,emp_id):
    for employee in self.employees:
      if employee.emp_id == emp_id:
        employee.display_info()

  def details_by_name(self,name):
    for employee in self.employees:
      if employee.name == name:
        employee.display_info()

  def details_all_employees(self):
    print("Check length",len(self.employees))
    for employee in self.employees:
        employee.display_info()
        print("====================")
```

```python
controller_object=EmployeeManagementSystem()
while True:
  print("Please choose the option")
  print("1) Add Employee")
  print("2) Delete Employee by ID")
  print("3) Update Name by ID ")
  print("4) Update Salary by ID")
  print("5) Update Job Position  by ID")
  print("6. Update age by ID ")
  print("7) Delete Employee by Name")
  print("8) Get details by ID")
  print("9) Get Details by Name")
  print("10) All details")
  choice=input("Please choose an option ")

  if choice=='1':
    name=choice=input("Please enter Employee name ")
    salary=choice=input("Please enter Employee salary ")
    age=choice=input("Please enter Employee age ")
    position=choice=input("Please enter Employee position ")
    id=len(controller_object.employees)
    emp_temp=Employee(id,name,position,salary,age)
    controller_object.add_employee(emp_temp)
  if choice=='2':
    id=int(input("Please enter Employee id you want to remove "))
    if controller_object.remove_employee(id):
      print("Employee Removed Successfully")
    else:
      print("Id not found could not remove the employee")

  if choice=='3':
    id=int(input("Please enter Employee ID you want to update "))
    new_name = input("Enter new name: ")
    if controller_object.update_name(id, new_name):
        print("Name updated successfully")
    else:
        print("Employee not found or could not update name")

  if choice=='4':
    id=int(input("Please enter Employee ID you want to update "))
    new_salary = float(input("Enter new salary: "))
    if controller_object.update_salary(id, new_salary):
        print("Salary updated successfully")
    else:
        print("Employee not found or could not update salary")

  if choice=='5':
    id=int(input("Please enter Employee ID you want to update "))
    new_position = input("Enter new job position: ")
    if controller_object.update_position(id, new_position):
        print("Position updated successfully")
    else:
        print("Employee not found or could not update position")

  if choice=='6':
    id=int(input("Please enter Employee ID you want to update "))
    new_age = int(input("Enter new age: "))
    if controller_object.update_age(id, new_age):
        print("Age updated successfully")
    else:
        print("Employee not found or could not update age")

  if choice=='7':
    name=input("Please enter Employee name you want to remove ")
    if controller_object.delete_by_name(name):
        print("Employee removed successfully")
    else:
        print("Employee not found or could not remove")

  if choice=='8':
    id=int(input("Please enter Employee ID you want to get details for "))
    controller_object.details_by_id(id)

  if choice=='9':
    name=input("Please enter Employee name you want to get details for ")
    controller_object.details_by_name(name)

  if choice=='10':
    controller_object.details_all_employees()

```
