```python
import random

# Function to display the Tic-Tac-Toe board
def display_board(board):
    print(" {} | {} | {} ".format(board[0], board[1], board[2]))
    print("-----------")
    print(" {} | {} | {} ".format(board[3], board[4], board[5]))
    print("-----------")
    print(" {} | {} | {} ".format(board[6], board[7], board[8]))
    print("\n")

# Function to check if a player has won
def check_winner(board, player):

    # Check rows and columns for a win
    for i in range(3):
        row_win = True
        column_win = True
        for j in range(3):
            if board[i*3 + j] != player:
                row_win = False
            if board[j*3 + i] != player:
                column_win = False
        if row_win or column_win:
            return True

    # Check diagonals for a win
    diagonal1_win = True
    for i in [0, 4, 8]:
        if board[i] != player:
            diagonal1_win = False

    diagonal2_win = True
    for i in [2, 4, 6]:
        if board[i] != player:
            diagonal2_win = False

    if diagonal1_win or diagonal2_win:
        return True

    return False

# Function to check if the board is full
def is_board_full(board):
    for cell in board:
        if cell == " ":
            return False
    return True

# Main game function
def play_tic_tac_toe():
    print("Welcome to Tic-Tac-Toe!")
    player_name = input("Enter your name: ")
    player_symbol = input("Hello, " + player_name + "! Choose X or O: ").upper()

    # Validate player's symbol choice
    while player_symbol not in ["X", "O"]:
        player_symbol = input("Please choose either X or O: ").upper()

    # Assign computer's symbol
    if player_symbol == "O":
        computer_symbol = "X"
    else:
        computer_symbol = "O"

    # Randomly decide who goes first
    if random.randint(0, 1) == 0:
        current_player = player_symbol
    else:
        current_player = computer_symbol

    print(current_player + " goes first!")

    # Initialize the game board
    board = [" " for _ in range(9)]

    while True:
        if current_player == player_symbol:
            # Player's turn
            position = int(input("Enter the index (1-9) where you want to place your symbol: ")) - 1

            # Validate the chosen position
            while board[position] != " " or position < 0 or position > 8:
                print("Invalid move. Try again.")
                position = int(input("Enter the index (1-9) where you want to place your symbol: ")) - 1
        else:
            # Computer's turn
            available_moves = []
            for i, cell in enumerate(board):
                if cell == " ":
                    available_moves.append(i)

            position = random.choice(available_moves)

        # Place the symbol on the board
        board[position] = current_player
        display_board(board)

        # Check for a winner or draw
        if check_winner(board, current_player):
            if current_player == player_symbol:
                print("Congratulations, you won!")
            else:
                print("Computer wins!")
            break
        elif is_board_full(board):
            print("It's a draw!")
            break

        # Switch players for the next turn
        if current_player == player_symbol:
            current_player = computer_symbol
        else:
            current_player = player_symbol

    # Ask if the player wants to play again
    play_again = input("Do you want to play again? (yes/no): ")
    if play_again.lower() == "yes":
        play_tic_tac_toe()
    else:
        print("Thank you for playing!")

# Start the game
play_tic_tac_toe()

```
