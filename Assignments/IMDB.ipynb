```python


# Define a function to extract time from the original "Run Time" column
def extract_run_time(value):
    if 'hour' in value and 'minute' in value:
        return value
    elif 'hour' in value:
        return value
    elif 'minute' in value:
        return value
    else:
        return 'N/A'

# Apply the function to extract time and create a new column "Actual Run Time"
df['Actual Run Time'] = df['Run Time'].apply(extract_run_time)

# Define a function to extract budget from the original "Run Time" column
def extract_budget(value):
    if 'estimated' in value.lower():
        return value
    elif any(currency_symbol in value for currency_symbol in ['$', 'â¬', 'Â£', 'Â¥', 'â¹']):
        return value
    else:
        return 'N/A'

# Apply the function to extract budget and create a new column "Budget"
df['Budget'] = df['Run Time'].apply(extract_budget)

# Clean the "Run Time" column by removing budget information
def clean_run_time(value):
    if 'hour' in value and 'minute' in value:
        return 'N/A'
    elif 'hour' in value:
        return 'N/A'
    elif 'minute' in value:
        return 'N/A'
    else:
        return value

df['Run Time'] = df['Run Time'].apply(clean_run_time)

# Drop the "Run Time" column
df = df.drop(columns=['Run Time'])
```

```python
df
```

```python
# Replace non-numeric characters with an empty string
df['year'] = df['year'].str.replace(r'\D', '', regex=True)

# Convert the 'year' column to numeric
df['year'] = pd.to_numeric(df['year'], errors='coerce')

# Filter out non-date values (assuming valid years are between 1900 and 2023)
df = df[(df['year'] >= 1900) & (df['year'] <= 2023)]
df['year'] = df['year'].astype('Int64')

# Reset the index
df.reset_index(drop=True, inplace=True)
```

```python

# Define a function to handle 'K' and 'M' in User Rating
def handle_k_m(value):
    if 'K' in value:
        return float(value.replace('K', '')) * 1000
    elif 'M' in value:
        return float(value.replace('M', '')) * 1000000
    return float(value)

# Apply the function to the User Rating column in the original DataFrame
df['User Rating'] = df['User Rating'].apply(handle_k_m)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of Ratings
plt.figure(figsize=(30, 9))
sns.histplot(df['Rating'], bins=20, kde=True)
plt.title('Distribution of Ratings')
plt.show()
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Define features and target variable
features = ['Budget']  # Add more features as needed
target = 'User Rating'

# Create a subset of the data with only relevant columns
subset = df[features + [target]].copy()

# Handle missing values and convert Budget to numeric
subset['Budget'] = pd.to_numeric(subset['Budget'].str.replace('[^\d]', '', regex=True), errors='coerce')
subset['Budget'].fillna(0, inplace=True)



# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(subset[features], subset[target], test_size=0.2, random_state=42)

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
